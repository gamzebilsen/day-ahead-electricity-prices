# Introduction
Energy Price Forecasting is a growing discipline, especially given a significant change in the energy grid with the current energy transition of higher renewables and demand response programs acceleratingly integrating. Wholesale market participants, buyers and sellers of energy for the independent energy operators to meet the demand with enough supply all hours of the day, need to be able to accurately predict prices. This prediction can occur months before by both parties such as a large generator (seller) and a town that needs to meet the resident’s electric needs (buyer) which is the futures market; both parties need to have an accurate understanding of what the prices might be to incorporate these prices and risk of changes in prices to be prepared for any significant changes in demand and supply that might impact the reliability of the grid, such as the cold snap that occurred in Texas in 2020, thus to hedge properly against risk. There’s also the day-ahead market, which this project focuses on, in which participants buy and sell energy for the next day. Based on their predictions of the prices, generators might find it too expensive to turn on a plant, leading to their not bidding in/participating in the market next day. Lastly, there’s the spot market, in which the supply sides enter the market to cover any last-minute discrepancies in the energy balance. 

This project as stated before focuses on the day-ahead market. Although futures market and spot prices are highly important, due to availability of data I could do either day-ahead or spot market. Given the supply and demand side impact on the day-ahead market compared to spot markets sole focus on the supply side as it stands, although the demand response programs are likely to adjust this in the future, I’ve chosen the day-ahead market price forecasting. 

Accurately predicting the energy prices of a day-ahead has significant financial impact on participants and most importantly on the wellbeing of the energy system. For a day-ahead markets, if a generator predicts lower prices than the actual thus decide to not participate and consequently turn on a given generator due to higher cost of starting it on the next day, firstly loses on gains they could’ve had. Most importantly however, the next day, if there are higher prices than predicted, in any case likely the supply fell short, leading to more activity in the spot market. Independent system operators who manage the reliability of the system and consequently manage the wholesale market then need to ask generators who are able to start off really quickly to meet this demand last minute – these plants that can quickly turn on unfortunately also tend to be the highest emitters. Thus, mistaken predictions can lead to lower emitting plants to not enter the market, causing both a threat to the reliability of the system – which is what occurred in Texas when the plants couldn’t turn on due to frozen pipes – and the most likely higher emissions if the supply ended up falling short due to the price discrepancy.

## Merging Datasets

```python
trans_constraints = pd.DataFrame()
demand_bids = pd.DataFrame()
load_forecast = pd.DataFrame()
for file in os.listdir():
    if file.startswith('da_trans'):
        x = pd.read_csv(file)
        trans_constraints = trans_constraints.append(x)
    if file.startswith('hrl_da_de'):
        x = pd.read_csv(file)
        demand_bids = demand_bids.append(x)
    if file.startswith('load_fr'):
        x = pd.read_csv(file)
        load_forecast = load_forecast.append(x)
```

```python
trans_constraints['datetime_beginning_ept'] = pd.to_datetime(trans_constraints['datetime_beginning_ept'])
trans_constraints['datetime_ending_ept'] = pd.to_datetime(trans_constraints['datetime_ending_ept'])
demand_bids['datetime_beginning_ept'] = pd.to_datetime(demand_bids['datetime_beginning_ept'])
load_forecast['forecast_hour_beginning_ept'] = pd.to_datetime(load_forecast['forecast_hour_beginning_ept'])
```

```python
trans_constraints = trans_constraints.drop(['datetime_beginning_utc','datetime_ending_utc'],axis=1)
load_forecast = load_forecast.drop(['evaluated_at_utc','evaluated_at_ept','forecast_hour_beginning_utc','forecast_area'],axis=1)
demand_bids = demand_bids.drop(['datetime_beginning_utc'],axis=1)
```

```python
demand_bids2 = demand_bids.groupby('datetime_beginning_ept').sum()
demand_bids2['$/MWh'] = demand_bids2['price_point']/demand_bids2['mw']
demand_bids3 = demand_bids2.drop(['price_point','mw'],axis=1)
load_forecast2 = load_forecast.groupby('forecast_hour_beginning_ept').sum()
```

Duration in trans_contraints is in terms of hours. 

```python
trans_time = trans_constraints.groupby('datetime_beginning_ept').mean().index
demand_time = demand_bids3.index
date_toappend = pd.DataFrame(list(set(demand_time) - set(trans_time))).rename({0:'datetime_beginning_ept'},axis=1)
trans_constraints['frequency'] = 1
date_toappend['frequency'] = 0
trans_constraints2 = pd.concat([trans_constraints,date_toappend])
trans_constraints2 = trans_constraints2.sort_values('datetime_beginning_ept')
trans_constraints2['duration'] = trans_constraints2['duration'].fillna(0)
trans_constraints3 = trans_constraints2[trans_constraints2['datetime_beginning_ept']>'2014-05-23 00:00:00']
```

```python
duration_list = list(trans_constraints3['duration'].value_counts().index)[1:]
duration_list = duration_list[:4]+duration_list[5:]
```

Different types of maintenances likely impact the prices more - say if it's a generator that covers most of the electricity during peak hour times - however, due to time constraints I will assume them to have equal impact thus add up frequencies per hour based on each maintainance duration.

Below, doing a very brute force of adding frequency of maintainence per hour. 

```python
for i in range(trans_constraints3.shape[0]):
    for duration in duration_list:  
        if trans_constraints3.iloc[i,2] == duration:
            column = trans_constraints3['datetime_beginning_ept'] 
            list_of = np.arange(duration)[1:]
            for number in list_of:
                number=int(number)
                equals  = trans_constraints3.iloc[i+number,1] + timedelta(hours=number)
                to_addvalue = trans_constraints3.iloc[i+number,-1]
                trans_constraints3.loc[trans_constraints3['datetime_beginning_ept'] == equals, "frequency"] = to_addvalue + 1
```

```python
trans_constraints4 = trans_constraints3.groupby('datetime_beginning_ept').sum()
```

```python
merge1 = pd.merge(load_forecast2,demand_bids3,right_index=True,left_index=True)
finaldata = pd.merge(merge1,trans_constraints4,right_index=True,left_index=True)
finaldata = finaldata[['forecast_load_mw','$/MWh','frequency']]
finaldata = finaldata[['forecast_load_mw','$/MWh','frequency']]
# finaldata.to_csv('finaldata.csv')
```

# Data description
The original dataset that I have are historical day-ahead prices, load-forecast and transmission constraint data by the PJM from April 2014 to April 2022. PJM is the Independent System Operator for many states including Pennsylvania and New Jersey and also includes parts of states such as Illinois, thus spans many states. 

For day-ahead prices, both of the papers cited in this project mention the main inputs to be one, two, three and seven day ahead prices and load forecast which I have added to the model. Jedrzejewski et al also recommends including solar and wind generation data given low marginal cost of these increasingly prevalent generation methods which is changing the structure thus the prices of the market. However, because to PJM’s data for wind and solar generation start in 2020, I didn’t include the data due to high likelihood of the introduction of data unavailability prior to 2020, given the states had wind and solar participation in the wholesale market before that data. I’ve also wanted to include transmission constraints as an input although it wasn’t recommended by any of the papers, because I thought less availability of plants in a given day due to construction likely increases prices, by calculating the frequency of constraints introduced by generators going offline in a given day. However, as can be seen in the code below, this ended up leading to unbalanced data with the number of offline generators climbing then falling. I assumed this is due to data error given more equalized levels of frequency after mid-2017, thus removed this from the final dataset. 

Then, I’ve looked at the annual and hourly distribution of the both price and load forecasting data to understand the seasonality and the trend to see if I can provide additional information to the model. In the time series graphs below, I’ve noticed a slight change in the trend around when Covid became more mainstream, so I’ve added a dummy variable for days after Covid. I’ve also added dummy variables for each day of the week, month and year to cover changes in consumption throughout the week - more electricity is consumed during the weekdays, months – more consumption during the summer months due to A/C, and years – more integration of renewables into the grid and increased energy efficiency throughout years. Lastly, I’ve also added US holiday dates as a dummy variable as electricity consumption tends to be lower on these days.

After finalizing the data, I scale the non-binary data with the min-max scaling to make all of the data be in between 0 and 1 including. I then separate data into train, validation and test datasets. Based on Lago et al’s recommendation, I wanted to use more than a year as the testing variable, and use close to a year as the validation period. Given my longer data which is 7.37 years old compared to their 5 years of data, I proportionately wanted to separate the data, thus have 1.5 years end of the data be in the testing period, and almost 11 months to be in the validation data. 

So, the <b> target variable</b> is the day-ahead electricity prices in the PJM region in $/MWH. The <b>features</b> are 1,2,3,7 days ago electricity prices and the day afters load forecast in addition to 1,2,3,7 days ago's load forecast. Lastly, there are features that are dummy variables to identify time aspect of the data such as day of week, month, year in question, if the date is a holiday and if the date is after covid.


# Why and How Deep Learning
There’s extensive literature to tackle the electric price forecasting problem that has been growing more so in the recent years. In the literature, most recently the LEAR model has been popularized which is a form of Lasso regression: given many inputs used to predict the prices, Lasso’s automatic feature extraction model has been proven to improve predictive capabilities given many variables input into the models as mentioned such as the 1,2,3,7 days before data combined with the dummy variables. On top of LEAR, with the popularization of neural networks due to higher accessibility to improved computation, deep learning has also been applied to predicting energy market information. Both Jedrzejewski et al and Lago et al highlight that although DNN’s improve performance compared to LEAR, more complex models that are widely used for EPF such as LSTM haven’t been proven to improve predictive capacity. 

Consequently, given their recommendation of using Deep Neural Networks, I will be doing so as well. However, on top of their recommendation based on the previous literature, I believe deep learning would be a great fit given availability of large datasets for EPF – my own data has above 60 thousand observations, and deep learnings best performs with larg datasets. In addition, deep learning allows us to use computers to find patterns beyond linear relationships unlike the LEAR model which is a type of regression that focuses on linear relationships. Third, it automates feature engineering thus on top of the little engineering that I’ve done, I don’t have to do more complex feature engineering nor need much more domain knowledge since deep learning is able to extract patterns that provide the most valuable information to predict the outcome easily using backpropagation and gradient descent: deep learning is able to employ both many regressions and non-linear relationships by using non-linear activation functions to change the weights in each mini-linear function within each layer and neuron, in order to minimize the loss function which is a function that has the prediction and the actual values as its inputs. However, this complex minimization of the loss function can have downsides which is firstly that the model can have vanishing or exploding gradients, thus the inability for the model to reach the optimal minimizing point in the gradient descent. Many alterations can be done to address this problem such as initializing weights – I’ve used Lecun in my examples, trying different activation functions, batch normalization or gradient clipping, thus it’s not a problem that doesn’t necessarily have solutions yet is still an issue. Additionally, the many complex functions deep learning intrinsically contains leads the models to have many parameters, leading to longer computation time than other machine learning models such as LEAR which is a major downside. However given the papers’ indication that DNN’ tend to outperform LEAR, it’s likely not much of an issue given the high cost of not predicting electricity prices well. Lastly, similar to some machine learning models such as SVM, deep learning doesn’t allow interpretability of the results thus we can’t necessarily know the interactions between the predictors and the outcome well after the model is ran unlike regression models like Lasso. 

However as stated, the ability for Deep Learning to combine linear and non-linear functions to find the optimal loss by employing backpropagation in gradient descent tends to allow it to have the best predictive capabilities. As stated, given the gravity of the problem of meeting electricity demand and supply accurately to not have an issue like it happened in Texas in 2020 when people went without electricity in the cold for days – despite the longer computation time and lack of interpretability, modelers can test out different methods to provide means for the model to reach the lowest loss thus to have the best predictions, which is the premise of this project. 

Next, you can initially find deep neural network tests, starting with simple neural networks to stacked networks, combination with different methods such as dropout and batch normalization to help improve the model’s predictive capabilities based on the training and validation loss combined with the performance of predictions compared to actual values using visualizations. 

After DNN models, I’ve tested different RNN models including simple GRU, LSTM and RNN. I’ve then tested a stacked RNN thus a variation of RNN and stacked LSTM. For variations of the LSTM models, I’ve reprocessed the data, thus used data reprocessing as an optimization.

In all of the models, I’ve decided to use mean squared error as the loss function given it’s a regression and not mean absolute error because the data doesn’t have much outliers given the long-time span. I’ve decided to use stochastic gradient descent as the optimizer for all as it seems to outperform other optimizers based on Geron’s recommendations. Testing different loss functions such as mean absolute error or optimizers such as RMSProp or NAG among many others in the future to improve the models is recommended. 

For DNN, I’ve tested 3 layers in total for DNN and Simple RNN and 2 layers for LSTM, however different number of layers and specific neurons per each layer is also recommended. I’ve also tested two levels of dropout for DNN and only one for both Simple RNN and LSTM, and a different learning rate for DNN, however different levels for each and other types of regularization such as l1 and l2 is recommended. Additionally, I’ve compared relu and selu activation functions for DNN and LSTM and used the default tanh for the simple RNN models, however different tests can be done given many different types of activation functions, and combined with different weight initializations such as Lecun or He among many others. Next, there are other types of regularizations such as monte carlo dropout or max-norm regularization that could have been employed to make sure the testing data is more accurately predicted, which I didn’t in this project. Lastly, different levels of gradient clipping and batch normalization can also be tested.


```python
#Preliminary function and reused values 
def history_plot(model_history,topylim,title):
    pd.DataFrame(model_history.history).plot(figsize=(15, 5)) 
    plt.grid(True)
    plt.gca().set_ylim(0, topylim)
    plt.title(title)
def to_graph_df(model):
    train_predict = pd.DataFrame(model.predict(X_train))#.rename({0:'predicted_train'},axis=1)
    train_predict['predict_type'] = 'train'
    test_predict = pd.DataFrame(model.predict(X_test))#.rename({0:'predicted_test'},axis=1)
    test_predict['predict_type'] = 'test'
    val_predict = pd.DataFrame(model.predict(X_val))#.rename({0:'predicted_validation'},axis=1)
    val_predict['predict_type'] = 'val'
    actual_data = pd.concat([pd.DataFrame(y_train),pd.DataFrame(y_val),pd.DataFrame(y_test)],axis=0).rename({'$/MWh':'actual_y'},axis=1)
    train_predict.index = X_train.index
    test_predict.index = X_test.index
    val_predict.index = X_val.index
    predicted_data = pd.concat([train_predict,test_predict,val_predict],axis=0)
    modeldata = predicted_data.merge(actual_data,left_index=True,right_index=True).rename({0:'predicted_y'},axis=1)
    return modeldata
def graph_results(model_data):
    ax = model_data[model_data['predict_type']=='test']['predicted_y'].plot(figsize=(15,7.5),color='red')
    ax = model_data[model_data['predict_type']=='train']['predicted_y'].plot(figsize=(15,7.5),color='blue')
    ax = model_data[model_data['predict_type']=='val']['predicted_y'].plot(figsize=(15,7.5),color='orange')
    ax = model_data['actual_y'].plot(figsize=(15,7.5),alpha=0.5)
    legend_elements = [Line2D([0], [0], color='blue', lw=3, label='Train'),
                       Line2D([0], [0], color='orange', lw=3, label='Validation'),
                       Line2D([0], [0], color='red', lw=3, label='Test'),
                       Line2D([0], [0], lw=3, label='Actual')
                       ]
    ax.legend(handles=legend_elements, loc='upper right')
def print_last_loss(history_input,title):
    print(f"{title} \nLoss: {history_input.history['loss'][-1]} \nValidation Loss: {history_input.history['val_loss'][-1]}")
EPOCHS = 100 
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
```


# Deep Learning Models
### DNN Model 1 - Simple DNN with Early Stopping

```python
model1_1 = tf.keras.models.Sequential([
         tf.keras.layers.Dense(30, activation="relu", input_shape=X_train.shape[1:]),
         tf.keras.layers.Dense(1)
     ])
model1_1.compile(loss="mean_squared_error", optimizer="sgd")

history1_1 = model1_1.fit(X_train, y_train, epochs=EPOCHS, callbacks=[early_stopping_cb],
         validation_data=(X_val, y_val),verbose=False)
```

### Code prep for RNN
Making the time steps be a month: the timeframe has x number of samples such that each samples the past thirty days as the input.

```python
#code from https://machinelearningmastery.com/how-to-develop-lstm-models-for-time-series-forecasting/
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        end_ix = i + n_steps
        if end_ix > len(sequence)-1:
            break
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
rnn_X_train = split_sequence(np.array(X_train),24*30)[0]
rnn_y_train = split_sequence(np.array(y_train),24*30)[0]
rnn_X_val = split_sequence(np.array(X_val),24*30)[0]
rnn_y_val = split_sequence(np.array(y_val),24*30)[0]
rnn_X_test = split_sequence(np.array(X_test),24*30)[0]
rnn_y_test = split_sequence(np.array(y_test),24*30)[0]
```

```python
def to_graph_dfrnn(model):
    train_predict = pd.DataFrame(model.predict(rnn_X_train))#.rename({0:'predicted_train'},axis=1)
    train_predict['predict_type'] = 'train'
    test_predict = pd.DataFrame(model.predict(rnn_X_test))#.rename({0:'predicted_test'},axis=1)
    test_predict['predict_type'] = 'test'
    val_predict = pd.DataFrame(model.predict(rnn_X_val))#.rename({0:'predicted_validation'},axis=1)
    val_predict['predict_type'] = 'val'
    actual_data = pd.concat([pd.DataFrame(y_train).iloc[720:,:],pd.DataFrame(y_val).iloc[720:,:],pd.DataFrame(y_test).iloc[720:,:]],axis=0).rename({'$/MWh':'actual_y'},axis=1)
    train_predict.index = X_train.iloc[720:,:].index
    test_predict.index = X_test.iloc[720:,:].index
    val_predict.index = X_val.iloc[720:,:].index
    predicted_data = pd.concat([train_predict,test_predict,val_predict],axis=0)
    actual_data.index=predicted_data.index
    modeldata = predicted_data.merge(actual_data,left_index=True,right_index=True).rename({0:'predicted_y'},axis=1)
    return modeldata
```

### RNN Model 4 - RNN with 3 layers, layer normalization, dropout and gradient clipping
The kernel started to die after I started running LSTM as it was the best model; because of this and long timespan of the simple models running, I will use RNN and conduct the recommended optimizations by Geron ='s book to see how it performs on RNN rather than making additions and finding the best combinations step by step like I've done for DNN.

Because RNN's tend to have gradient problems, I will use both gradient clipping and layer normalization and dropout (rather than testing them one by one) to see how it performs, while using selu activation assuming it will outperform the default tanh and the relu.

I will use 0.3 dropout compared to the DNN's lower dropout performance bc the simple Simple RNN predicted with very low variance, possibly leading to overfitting issues.

I will use 0.5 as the clipping value given the recommendation is usually to be between -1 and 1.

```python
optimizer = tf.keras.optimizers.SGD(clipnorm=0.5)
initializer = tf.keras.initializers.LecunNormal()

#Below code is from Geron Chapter 15
class LNSimpleRNNCell(tf.keras.layers.Layer):
    def __init__(self, units, activation="selu", **kwargs):
        super().__init__(**kwargs)
        self.state_size = units
        self.output_size = units
        self.simple_rnn_cell = tf.keras.layers.SimpleRNNCell(units,activation=None,kernel_initializer=initializer)
        self.layer_norm = tf.keras.layers.LayerNormalization()
        self.activation = tf.keras.activations.get(activation) 
    def call(self, inputs, states):
        outputs, new_states = self.simple_rnn_cell(inputs, states) 
        norm_outputs = self.activation(self.layer_norm(outputs)) 
        return norm_outputs, [norm_outputs]
    
model2_4 = tf.keras.models.Sequential([
        tf.keras.layers.RNN(LNSimpleRNNCell(20), return_sequences=True,input_shape=(720,38)),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.RNN(LNSimpleRNNCell(20), return_sequences=True),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.RNN(LNSimpleRNNCell(10), return_sequences=False),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(1)
    ])

model2_4.compile(loss="mean_squared_error", optimizer=optimizer)

history2_4 = model2_4.fit(rnn_X_train, rnn_y_train, epochs=10, 
                         validation_data = (rnn_X_val, rnn_y_val),verbose=False)
```

```python
history_plot(history2_4,0.01,'Complex RNN')
```

```python
# example of calculate the root mean squared error
from sklearn.metrics import mean_squared_error
def test_mse_calc(model):
    test_predict = pd.DataFrame(model.predict(X_test))
    test_predict.index = X_test.index
    modeldata = test_predict.merge(y_test,left_index=True,right_index=True).rename({0:'predicted_y'},axis=1)
    errors = mean_squared_error(modeldata['$/MWh'], modeldata['predicted_y'], squared=False)
    return errors
```
